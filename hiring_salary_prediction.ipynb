```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("data/hiring.csv")
```

```python
#Confirming proper load
print("First 5 rows of the dataset:")
print(df.head())
```

```python
#Cleaning the data, handling missing values and fixing the experience column from text to numeric

experience_map = {
    "zero": 0,
    "one": 1,
    "two": 2,
    "three": 3,
    "four": 4,
    "five": 5,
    "six": 6,
    "seven": 7,
    "eight": 8,
    "nine": 9,
    "ten": 10,
    "eleven": 11,
    "twelve": 12
}

df["experience"] = df["experience"].replace(experience_map)

df.fillna(0, inplace=True)

df["experience"] = df["experience"].astype(int)

print("\nDataset after cleaning:")
print(df)
```

```python
#Splitting the cleaned data to features (X) and targets(Y)
# Features = the input columns we use to predict salary.
# Target = the column we want to predict.

X = df[["experience", "test_score(out of 10)", "interview_score(out of 10)"]]
y = df["salary($)"]

print("\nFeatures (X) preview:")
print(X.head())
print("\nTarget (y) preview:")
print(y.head())
```

```python
#Create and train the model (Linear regression model)

model = LinearRegression()
model.fit(X, y)

print("\nModel trained successfully!")
```

```python
#Now we predict salaries for new hirers
# Candidate 1: 2 yrs experience, 9 test score, 6 interview score
# Candidate 2: 12 yrs experience, 10 test score, 10 interview score

predictions = model.predict([
    [2, 9, 6],
    [12, 10, 10]
])

print("\nPredicted Salaries:")
print("Candidate 1 (2 yrs, 9 test, 6 interview):", predictions[0])
print("Candidate 2 (12 yrs, 10 test, 10 interview):", predictions[1])
```
